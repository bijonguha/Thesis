```python
def greyscale(img):
    if len(img.shape) == 3 and img.shape[2] == 3:
        weight = [0.2,0.5,0.3] # weights
        weighted_mean = np.tensordot(img,weight, axes=((-1,-1)))[...,None]
    return(weighted_mean)
```

```python
def add_data(data, spectrum, class_name, sample_no):
    # add segment fft to data_detailed
    new_sample = np.array(spectrum.reshape(spectrum.size).tolist())
    new_sample = new_sample.reshape(spectrum.size, 1)
    new_df = pd.DataFrame(new_sample.T, columns=df_header[:-1])
    new_df['class'] = class_name

    if data.empty:
        data = new_df
    else:
        data = data.append(new_df, ignore_index=True)

    return data
```

```python
import shutil

def save_data(df_all_segments, class_name, image_count, image):
    #if image_count%2 == 0 or image_count == 0: #train set
    if not (image_count == 8 or image_count == 10): #train set
        train_file_path = train_folder + class_name + "_train.pkl"
        if not image_count == 0:
            train_unpickled_df = pd.read_pickle(train_file_path)
            df_all_segments = train_unpickled_df.append(df_all_segments, ignore_index=True)
        df_all_segments.to_pickle(train_file_path)
    else: #test set
        test_file_path = test_folder + class_name + "_test.pkl"
        if not image_count == 8:
            test_unpickled_df = pd.read_pickle(test_file_path)
            df_all_segments = test_unpickled_df.append(df_all_segments, ignore_index=True) 
        df_all_segments.to_pickle(test_file_path)
        print(class_name, ":", image_count)
        plt.imshow(image)
        plt.show()
```

```python
def darken_region(magnitude_spectrum):
    # darken region around axis
    middle = int(magnitude_spectrum.shape[0] / 2)
    magnitude_spectrum[middle - 5:middle + 5, :] = magnitude_spectrum.min()
    magnitude_spectrum[:, middle - 5:middle + 5] = magnitude_spectrum.min()
    return(magnitude_spectrum)
```

```python
def compute_accumulated_spectra(train, img_names, path, number_of_snippets, class_name):
    # number of pixels to move each segmentä
    data = pd.DataFrame()
    accumulated_spectra = np.zeros((SNIPPET_WIDTH, SNIPPET_WIDTH))
    step = np.empty([2], dtype=int)
    step[0] = SNIPPET_WIDTH
    step[1] = SNIPPET_WIDTH

    segment_count = np.empty([2], dtype=int)

    image_count = 0

    overall_sample_no = 0
    testsegmentcount = 0
    testimagescount = 0

    for image in img_names:
        img_path = path + '/' + image
        img = imageio.imread(img_path)

        grey_img = greyscale(img)

        # number of segments in one scanned image (in both dimensions)
        segment_count[0] = grey_img.shape[0] // step[0]
        segment_count[1] = grey_img.shape[1] // step[1]

        sample_no = 0

        # for(each segment in img): cut of segment
        for i in range(0, segment_count[0]):
            for j in range(0, segment_count[1]):
                # segment as copy of img snippet
                start_i = i * SNIPPET_WIDTH
                start_j = j * SNIPPET_WIDTH
                segment = grey_img[start_i:(start_i + SNIPPET_WIDTH), start_j:(start_j + SNIPPET_WIDTH), 0]

                window = signal.windows.hann(segment.shape[0]) #Hann window .. why?
                #window = signal.windows.bartlett(segment.shape[0])
                #plt.plot(segment*window*window.T)
                #plt.show()

                # fft with scipy
                #spectrum = fftpack.fft2(segment)
                #shift = fftpack.fftshift(spectrum) # move zero frequency component to the center
                #spectrum = 20 * np.log( np.abs(shift))
                
                # fft with opencv
                dft = cv2.dft(np.float32(segment*window*window.T),flags = cv2.DFT_COMPLEX_OUTPUT)
                dft_shift = np.fft.fftshift(dft)

                spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))

                spectrum = darken_region(spectrum) #.. why?
                
                
                #plt.subplot(121),plt.imshow(segment, cmap = 'gray')
                #plt.title('Input Image '+class_name), plt.xticks([]), plt.yticks([])
                #plt.subplot(122),plt.imshow(spectrum, cmap = 'gray')
                #plt.title('Magnitude Spectrum cv2 '+class_name), plt.xticks([]), plt.yticks([])
                #plt.show()

                #plt.subplot(221),plt.imshow(segment, cmap = 'gray')
                #plt.title('Input Image'), plt.xticks([]), plt.yticks([])
                #plt.subplot(222),plt.imshow(spectrum, cmap = 'gray')
                #plt.title('Magnitude Spectrum darkened'), plt.xticks([]), plt.yticks([])
                #plt.show()

                # Spektrum/number_of_snippets -> comparability
                mag_shift = np.divide(spectrum, number_of_snippets) 

                #if image_count%2 == 0 or image_count == 0: #train set
                if not (image_count == 8 or image_count == 10): #train set
                    # compute the accumulated fingerprint
                    accumulated_spectra = np.add(accumulated_spectra, mag_shift)

                # save sampled spectra for training analysis later
                data = add_data(data, spectrum, class_name, sample_no)
                sample_no += 1
                overall_sample_no += 1
        save_data(data, class_name, image_count, img)
        image_count+=1
        data = pd.DataFrame()
    print("Gesamtanzahl Samples:", overall_sample_no)
    print("Gesamtanzahl testsamples:", testsegmentcount)
    print("Gesamtanzahl testbilder:", testimagescount)
        
    return accumulated_spectra
```

```python
 def save_fingerprint(magnitude_all, class_name, train):

        # add class and add one per printer to pandas dataframes
        # norm additive + multiplied spectras
        # magnitude_all = np.divide(magnitude_all, magnitude_all.max())
        # magnitude_all = magnitude_all - magnitude_all.mean()
        # magnitude_all = np.divide(magnitude_all, (magnitude_all.max()-magnitude_all.min()))

        # norm the accumulated_spectra
        idx = np.argsort(magnitude_all.flatten())
        b = magnitude_all.flatten()[idx]

        # feature count can only be as large as snippet pixels amount
        #TODO: return message to inform user
        if b.size <= number_pixels:
            features = b.size - 1
        else:
            features = b.size - number_pixels

        threshold = b[features]

        mini = magnitude_all.min()
        # select only 1000 brightest pixel
        # TODO: better feature selection
        print("shape mag",magnitude_all.shape)
        #magnitude_all = [mag if not (math.isnan(mag) or math.isinf(mag)) else mini for mag in magnitude_all ]
        for i, mag in enumerate(magnitude_all):
            for j, mag in enumerate(mag):
                if math.isnan(mag) or math.isinf(mag):
                    magnitude_all[i][j] = mini  
        magnitude_all[magnitude_all < threshold] = mini
        # magnitude_all_multi[magnitude_all_multi<threshold] = mini

        magnitude_all = magnitude_all - mini
        magnitude_all = magnitude_all / magnitude_all.sum() # divide by sum, to make fingerprints comparable

        feature_vector = np.array(magnitude_all.reshape(magnitude_all.size).tolist() + [class_name])
        feature_vector = feature_vector.reshape(magnitude_all.size+1, 1)
        fingerprint_df = pd.DataFrame(feature_vector.T, columns=df_header)

        # save image: merged frequency spectrum by addition
        f_add = Image.fromarray(np.divide(magnitude_all, magnitude_all.max()) * 255).convert('RGB')

        # save image: merged frequency spectrum by addition
        path = SUBPATH + '/fingerprint/'

        if not os.path.exists(path):
            os.makedirs(path)

        if train:
            img_path = path + class_name + "_fingerprint.png"
            data_path = path + class_name + "_fingerprint.pkl"
        else:
            img_path = path + "test_" + class_name + "_fingerprint.png"
            data_path = path + "test_" + class_name + "_fingerprint.pkl"

        f_add.save(img_path, "PNG")

        magnitude_df = pd.DataFrame(magnitude_all.reshape(magnitude_all.size).tolist() + [class_name])

        magnitude_df.to_pickle(data_path)

        fingerprint_png = f_add
        accumulated_spectra = magnitude_all
        
        return fingerprint_png, accumulated_spectra
```

```python
def feature_extractor(path, img_names, number_of_snippets, class_name, train=True):
    accumulated_spectra = compute_accumulated_spectra(train, img_names, path, number_of_snippets, class_name)
    fingerprint_png, accumulated_spectra = save_fingerprint(accumulated_spectra, class_name, train)
    return accumulated_spectra
```

```python
def cal_fingerprint(path, classes):
    class_name = os.path.split(path)[-1]
    number_of_snippets = 0
    img_names = os.listdir(path)
    img_names = [image for image in img_names if not image.startswith('.')]
    img_names.sort()
    number_of_images = len(img_names)

    # load images
    for image in img_names: 
        img_path = path + '/' + image
        img = Image.open(img_path)
        width, height = img.size
        number_of_snippets_x_axis = width // SNIPPET_WIDTH
        number_of_snippets_y_axis = height // SNIPPET_WIDTH
        print(number_of_snippets_x_axis*number_of_snippets_y_axis)
        number_of_snippets += number_of_snippets_x_axis * number_of_snippets_y_axis
        
    fingerprint = feature_extractor(path, img_names, number_of_snippets, class_name)
    print("number_of_snippets:", number_of_snippets)
    
    return fingerprint, number_of_images, number_of_snippets
   
```

```python
from scipy.spatial import distance

def compare_to_fingerprints(sample):
    gtd = sample['class']
    prediction = 0
    prediction_value = 0
    passed = False
    amount_classes = len(classes)
    correlation_values = [0 for i in range(amount_classes)]

    for i in range(amount_classes):
        fi_print = fingerprints[i]
        fi_img = Image.fromarray(np.divide(fi_print, fi_print.max()) * 255).convert('RGB')
        #plt.subplot(211),plt.imshow(fi_img, cmap = 'gray')
        #plt.title('fingerprint'), plt.xticks([]), plt.yticks([])
        #plt.show()
        #fi_print_normed = fi_print / fi_print.sum()
        fi_print_sample = np.array(sample)[:-1].reshape(SNIPPET_WIDTH,SNIPPET_WIDTH).astype(np.float64)
        fi_sample_img = Image.fromarray(np.divide(fi_print_sample, fi_print_sample.max()) * 255).convert('RGB')
        #plt.subplot(212),plt.imshow(fi_sample_img, cmap = 'gray')
        #plt.title('sample spectrum:'), plt.xticks([]), plt.yticks([])
        #plt.show()
        
        # np.dot: Matrix multiplication --> correlation between fingerprint and sample
        ## kein richtiges Ähnlichkeitsmaß?
        correlation_values[i] = np.dot(fi_print_sample.reshape(fi_print_sample.size),
                                       fi_print.reshape(fi_print.size))
        
        #mal mit euklid versuchen.. --> führt zum selben Ergebnis
        #ed = distance.euclidean(fi_print_sample.reshape(fi_print_sample.size), fi_print.reshape(fi_print.size))
        #relative Distanz
        #correlation_values[i] = 1/(1+ed) 
        

        # prediction --> biggest correlation
        if correlation_values[i] > prediction_value:
            prediction_value = correlation_values[i]
            prediction = classes[i]

    if prediction == 0:
        print('ERROR: correlation values not correct.. still 0 as class!')

    if prediction == gtd:
        passed = True

    return(prediction, gtd, passed)
```

```python
result_df = pd.DataFrame()

for i in range(len(classes)):
    t_data = get_test_data(classes[i])
    
    # for every sample / snippet
    for index, row in t_data.iterrows():
        prediction, gtd, passed = compare_to_fingerprints(row) # return (predicted class, GTD, True or False prediction)

        result = [prediction, gtd, passed]

        if len(result_df) == 0:
            result_df = pd.DataFrame(np.array(result).reshape(1,3), columns=['prediction', 'GTD', 'passed'])
        else:
            result_df = result_df.append({'prediction': result[0], 'GTD': result[1], 'passed': result[2]},
                               ignore_index=True)
            
        if index > t_data.shape[0]:
            print("ERROR: with index in evaluation!!!!!")

display(result_df.T)
```

```python
import seaborn as sn

evaluation = result_df.replace({'passed': {'True': True, 'False': False}})

true_positives = []
true_negatives = []
false_positives = []
false_negatives = []
accuracy = []
precision = []
recall = []

for class_name in classes:
    class_only_in_GTD_df = evaluation[evaluation['GTD'] == class_name]
    class_not_in_GTD_df = evaluation[evaluation['GTD'] != class_name]
    tp = class_only_in_GTD_df[class_only_in_GTD_df['passed']==True]
    tp= tp.shape[0]
    tp_rate = tp / class_only_in_GTD_df.shape[0]
    tn = class_not_in_GTD_df[class_not_in_GTD_df['prediction']!=class_name]
    tn = tn.shape[0]
    tn_rate = tn / class_not_in_GTD_df.shape[0]
    fn = class_only_in_GTD_df[class_only_in_GTD_df['passed']==False]
    fn = fn.shape[0]
    fn_rate = fn / class_only_in_GTD_df.shape[0]
    fp = class_not_in_GTD_df[class_not_in_GTD_df['prediction']==class_name]
    fp = fp.shape[0]
    fp_rate = fp / class_not_in_GTD_df.shape[0]
    # accuracy = (TP + TN) / (P+N)
    a = (tp + tn) / (class_only_in_GTD_df.shape[0] + class_not_in_GTD_df.shape[0])
    if (tp+fn) > 0 and tp > 0:
        p = (tp) / (tp+fp)
    else:
        p = 0
    if (tp+fn) > 0:
        r = (tp) / (tp+fn)
    else:
        r = 0
    
    true_positives.append(tp)
    true_negatives.append(tn)
    false_positives.append(fp)
    false_negatives.append(fn)
    accuracy.append(round(a*100,1))
    precision.append(round(p*100,1))
    recall.append(round(r*100,1))

    # *100 to get % instead of rate
    #true_positives.append(round(tp_rate*100,1))
    #true_negatives.append(round(tn_rate*100,1))
    #false_positives.append(round(fp_rate*100,1))
    #false_negatives.append(round(fn_rate*100,1))
    #accuracy.append(round(a*100,1))
    
display(evaluation[evaluation['passed']==False])

statistics = [true_positives, true_negatives, false_positives, false_negatives, accuracy, precision, recall]

statistics = pd.DataFrame(np.array(statistics).reshape(7,12), columns=classes, index=["true_positives", "true_negatives", "false_positives", "false_negatives", "accuracy", "precision", "recall"])

display(statistics)

confusion_matrix = pd.DataFrame(np.zeros((12, 12)).reshape(12,12), columns=classes, index=classes)

for class_name1 in classes:
    for class_name2 in classes:
        gtds = evaluation[evaluation['GTD'] == class_name2]
        count = len(gtds[gtds['prediction'] == class_name1])
        confusion_matrix[class_name1][class_name2] = confusion_matrix[class_name1][class_name2]+count

display(confusion_matrix)
sn.color_palette("rocket_r", as_cmap=True)
sn.heatmap(confusion_matrix, annot=True, square=False, 
        linewidths=2, cmap='Blues', fmt='g')

plt.show()       
```
